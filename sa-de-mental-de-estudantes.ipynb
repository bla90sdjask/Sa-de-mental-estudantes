## 3. Distribuição de Idade:

```python
# Distribuição de Idade
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title('Distribuição de Idade dos Estudantes')
plt.xlabel('Idade')
plt.ylabel('Contagem')
plt.show()
```

## 4. Distribuição por Gênero:

```python
# Contagem de gênero
plt.figure(figsize=(6, 4))
sns.countplot(x='Choose your gender', data=df, palette='Set2')
plt.title('Distribuição por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.show()
```

## 5. Distribuição de Sintomas de Saúde Mental:

```python
# Distribuição de Depressão, Ansiedade e Ataques de Pânico
plt.figure(figsize=(10, 6))
sns.countplot(x='Do you have Depression?', data=df, palette='pastel')
plt.title('Estudantes com Depressão')
plt.show()

sns.countplot(x='Do you have Anxiety?', data=df, palette='muted')
plt.title('Estudantes com Ansiedade')
plt.show()

sns.countplot(x='Do you have Panic attack?', data=df, palette='coolwarm')
plt.title('Estudantes com Ataques de Pânico')
plt.show()
```

```python
# Calcular porcentagens de sintomas de saúde mental
total_estudantes = len(df)

# Depressão
depression_counts = df['Do you have Depression?'].value_counts()
depression_percentage = (depression_counts / total_estudantes) * 100

# Ansiedade
anxiety_counts = df['Do you have Anxiety?'].value_counts()
anxiety_percentage = (anxiety_counts / total_estudantes) * 100

# Ataques de Pânico
panic_attack_counts = df['Do you have Panic attack?'].value_counts()
panic_attack_percentage = (panic_attack_counts / total_estudantes) * 100

print(f"Porcentagem de estudantes com Depressão:\n{depression_percentage}")
print(f"\nPorcentagem de estudantes com Ansiedade:\n{anxiety_percentage}")
print(f"\nPorcentagem de estudantes com Ataques de Pânico:\n{panic_attack_percentage}")
```

## 6. Existe alguma correlação entre o gênero e a presença de sintomas de saúde mental?

```python
# Tabela de contingência para gênero e sintomas de saúde mental
gender_depression = pd.crosstab(df['Choose your gender'], df['Do you have Depression?'])
gender_anxiety = pd.crosstab(df['Choose your gender'], df['Do you have Anxiety?'])
gender_panic_attack = pd.crosstab(df['Choose your gender'], df['Do you have Panic attack?'])

print("Distribuição de Gênero e Depressão:\n", gender_depression)
print("\nDistribuição de Gênero e Ansiedade:\n", gender_anxiety)
print("\nDistribuição de Gênero e Ataques de Pânico:\n", gender_panic_attack)

# Gráfico para correlação entre gênero e sintomas de saúde mental
plt.figure(figsize=(10, 5))
sns.countplot(x='Choose your gender', hue='Do you have Depression?', data=df, palette='pastel')
plt.title('Distribuição de Depressão por Gênero')
plt.show()

sns.countplot(x='Choose your gender', hue='Do you have Anxiety?', data=df, palette='muted')
plt.title('Distribuição de Ansiedade por Gênero')
plt.show()

sns.countplot(x='Choose your gender', hue='Do you have Panic attack?', data=df, palette='coolwarm')
plt.title('Distribuição de Ataques de Pânico por Gênero')
plt.show()
```

##  7. Há uma correlação entre o ano de estudo e a presença de sintomas de saúde mental?

```python
# Tabela de contingência para ano de estudo e sintomas de saúde mental
year_depression = pd.crosstab(df['Your current year of Study'], df['Do you have Depression?'])
year_anxiety = pd.crosstab(df['Your current year of Study'], df['Do you have Anxiety?'])
year_panic_attack = pd.crosstab(df['Your current year of Study'], df['Do you have Panic attack?'])

print("Distribuição de Ano de Estudo e Depressão:\n", year_depression)
print("\nDistribuição de Ano de Estudo e Ansiedade:\n", year_anxiety)
print("\nDistribuição de Ano de Estudo e Ataques de Pânico:\n", year_panic_attack)

# Gráfico para correlação entre ano de estudo e sintomas de saúde mental
plt.figure(figsize=(10, 5))
sns.countplot(x='Your current year of Study', hue='Do you have Depression?', data=df, palette='pastel')
plt.title('Distribuição de Depressão por Ano de Estudo')
plt.show()

sns.countplot(x='Your current year of Study', hue='Do you have Anxiety?', data=df, palette='muted')
plt.title('Distribuição de Ansiedade por Ano de Estudo')
plt.show()

sns.countplot(x='Your current year of Study', hue='Do you have Panic attack?', data=df, palette='coolwarm')
plt.title('Distribuição de Ataques de Pânico por Ano de Estudo')
plt.show()
```

## 8. Estudantes casados versus solteiros: existe uma diferença na prevalência de sintomas de saúde mental?

```python
# Tabela de contingência para estado civil e sintomas de saúde mental
marital_depression = pd.crosstab(df['Marital status'], df['Do you have Depression?'])
marital_anxiety = pd.crosstab(df['Marital status'], df['Do you have Anxiety?'])
marital_panic_attack = pd.crosstab(df['Marital status'], df['Do you have Panic attack?'])

print("Distribuição de Estado Civil e Depressão:\n", marital_depression)
print("\nDistribuição de Estado Civil e Ansiedade:\n", marital_anxiety)
print("\nDistribuição de Estado Civil e Ataques de Pânico:\n", marital_panic_attack)

# Gráfico para correlação entre estado civil e sintomas de saúde mental
plt.figure(figsize=(10, 5))
sns.countplot(x='Marital status', hue='Do you have Depression?', data=df, palette='pastel')
plt.title('Distribuição de Depressão por Estado Civil')
plt.show()

sns.countplot(x='Marital status', hue='Do you have Anxiety?', data=df, palette='muted')
plt.title('Distribuição de Ansiedade por Estado Civil')
plt.show()

sns.countplot(x='Marital status', hue='Do you have Panic attack?', data=df, palette='coolwarm')
plt.title('Distribuição de Ataques de Pânico por Estado Civil')
plt.show()
```

## 9. Quantos estudantes que apresentam sintomas de saúde mental procuraram tratamento especializado?

```python
# Contagem de estudantes que procuraram tratamento especializado
treatment_needed = df[df['Do you have Depression?'] == 'Yes']['Did you seek any specialist for a treatment?'].value_counts()
print("Estudantes com Depressão que procuraram tratamento:\n", treatment_needed)

treatment_needed_anxiety = df[df['Do you have Anxiety?'] == 'Yes']['Did you seek any specialist for a treatment?'].value_counts()
print("\nEstudantes com Ansiedade que procuraram tratamento:\n", treatment_needed_anxiety)

treatment_needed_panic = df[df['Do you have Panic attack?'] == 'Yes']['Did you seek any specialist for a treatment?'].value_counts()
print("\nEstudantes com Ataques de Pânico que procuraram tratamento:\n", treatment_needed_panic)

# Gráfico para tratamento especializado
plt.figure(figsize=(10, 5))
sns.countplot(x='Did you seek any specialist for a treatment?', data=df[df['Do you have Depression?'] == 'Yes'], palette='pastel')
plt.title('Estudantes com Depressão que Procuraram Tratamento')
plt.show()

sns.countplot(x='Did you seek any specialist for a treatment?', data=df[df['Do you have Anxiety?'] == 'Yes'], palette='muted')
plt.title('Estudantes com Ansiedade que Procuraram Tratamento')
plt.show()

sns.countplot(x='Did you seek any specialist for a treatment?', data=df[df['Do you have Panic attack?'] == 'Yes'], palette='coolwarm')
plt.title('Estudantes com Ataques de Pânico que Procuraram Tratamento')
plt.show()
```
